```python
##接口函数

##计算点和直线的关系
def Point3_2(data, i, j, k):
    #print(i, data[i, :])
    #print(j, data[j, :])
    p1 = np.array([data[i, :]])
    p2 = np.array([data[j, :]])
    p3 = np.array([data[k, :]])
   # print(p1, p2, p3)
    k = (p1[0, 1] - p2[0, 1])/(p1[0, 0] - p2[0, 0])
    re = k * (p3[0, 0] - p2[0, 0]) + (p2[0, 1] - p3[0, 1])
    return re
def Point4Line_2(data, p1, p2, p3, p4):
    #i j k l
    #plot point
    #plt.figure('1')
    #plt.plot(X, Y)
    '''
    for idx,i in enumerate([p1, p2, p3, p4]):
        plt.scatter(i[0, 0], i[0, 1])
        plt.annotate(s = r'p%s' % str(idx+1), xy=(i[0, 0], i[0, 1]),xytext=(+20,-20), \
                     xycoords='data',textcoords='offset points', \
                     fontsize=16,arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2")) 
    plt.show()
    '''
    
    
    #select p1
    if Point3_2(data, p2, p3, p1)*Point3_2(data, p2, p3, p4)>=0 and Point3_2(data,p2, p4, p1)*Point3_2(data,p2, p4, p3)>=0 \
        and Point3_2(data,p3, p4, p1)*Point3_2(data,p3, p4, p2)>=0:
        #print('%f is in inner' % p1)
        data[p1, 2] = 0
        #return p2, p3, p4
        return 
    #select p2 
    if Point3_2(data,p1, p3, p2)*Point3_2(data,p1, p3, p4)>=0 and Point3_2(data,p1, p4, p2)*Point3_2(data,p1, p4, p3)>=0 \
        and Point3_2(data,p3, p4, p2)*Point3_2(data,p3, p4, p1)>=0:
        #print('%f is in inner' % p2)
        data[p2, 2] = 0
        #return p1, p3, p4
        return 
    #select p3
    if Point3_2(data,p1, p2, p3)*Point3_2(data,p1, p2, p4)>=0 and Point3_2(data,p1, p4, p3)*Point3_2(data,p1, p4, p2)>=0 \
        and Point3_2(data,p2, p4, p3)*Point3_2(data,p2, p4, p1)>=0:
        #print('%f is in inner' % p3)
        data[p3, 2] = 0
        #return p1, p2, p4
        return 
    #select p4
    if Point3_2(data,p1, p2, p4)*Point3_2(data,p1, p2, p3)>=0 and Point3_2(data,p1, p3, p4)*Point3_2(data,p1, p3, p2)>=0 \
        and Point3_2(data,p2, p3, p4)*Point3_2(data,p2, p3, p1)>=0:
        #print('%f is in inner' %p4)
        data[p4, 2] = 0
        #return p1, p2, p3
        return 
    return 
#print(p1, p2, p3, p4)


def Enume(data, N):
    for  p1 in range(N - 4):
        if data[p1, 2] == 0:
            continue
    #p1 = np.array([data[i, :]])
        for p2 in range(p1+1, N -3):
            if data[p2, 2] == 0:
                continue
            #p2 = np.array([data[j, :]])
            for p3 in range(p2+1, N-2):
                if data[p3, 2] == 0:
                    continue
                #p3 =  np.array([data[k, :]])
                for p4 in range(p3+1, N):
                    if data[p4, 2] == 0:
                        continue
                    #p4 = np.array([data[l, :]]) 
                    #print(p1, p2, p3, p4)
                    Point4Line_2(data, p1, p2, p3, p4)
                    #print(temp.shape)
                    #print(result.shape)
                    #print(result)
    result = []
    for i in range(N):
       # print(i, data[i, 2])
        if data[i, 2] == 1:
            result.append(data[i, :])
    result = np.array(result)
    #按照书上写法，进行逆时针输出
    xmin = xmax = 0
    for i in range(len(result) - 1):
        i = i + 1
        if result[xmin, 0] > result[i, 0]:
            xmin = i
        if result[xmax, 0] < result[i, 0]:
            xmax = i
    pmin = np.array([result[xmin, :]])
    pmax = np.array([result[xmax, :]])

    #print(pmin, pmax)
    Su = []
    Sl = []
    #print(result)
    for i in range(len(result)):
        if i != xmin and i!= xmax:
            #print(result[i,:])
            p = np.array([result[i,:]])
             
            if line2point(pmin, pmax, p) > 0:
                Su.append(result[i,:])
            else:
                Sl.append(result[i,:])

    #PrintResult = []
    #PrintResult.append(result[xmin, :])

    PrintResult = np.array([result[xmin, :]])
    #print(temp.shape)

    Su = np.array(Su)
    #print(Su)
    if len(Su) != 1 and Su != []:
        Su =Su[np.argsort(Su[:, 0])]
        PrintResult = np.concatenate((PrintResult, Su),axis=0)

    
    #PrintResult.append(Su)

    PrintResult = np.concatenate((PrintResult, np.array([result[xmax, :]])),axis=0)
    #PrintResult.append(result[xmax, :])

    #print(Sl)
    Sl = np.array(Sl)
    if len(Sl) != 1 and Sl != []:
        Sl = Sl[np.argsort(-Sl[:, 0])]
        PrintResult = np.concatenate((PrintResult, Sl),axis=0)
    
    return PrintResult
```

## 2. 实现基于Graham-Scan的凸包求解算法

```python
## 生成数据，寻找p0
import numpy as np
import matplotlib.pyplot as plt
L = 10
axis = np.random.randint(0, 100, (L, 2))
flag = np.ones((L, 3), dtype=np.int16)
Data = np.concatenate((axis, flag),axis=1)

#找纵坐标最大值
pidx = 0
for idx in range(L):
    if (Data[idx, 1] < Data[pidx, 1]):
        pidx = idx
ResData = np.delete(Data, pidx, axis = 0)
print(Data[pidx])
print(ResData)

#画图展示
#p0
plt.scatter(Data[pidx, 0], Data[pidx, 1])
plt.annotate(r'p0',xy=(Data[pidx, 0],Data[pidx, 1]), xycoords='data',\
xytext=(-20, +20), textcoords='offset points', fontsize=12,\
arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.1"))
#剩余的点
x = ResData[:, 0]
y = ResData[:, 1]
plt.scatter(x, y)
plt.show()
```

```python
##剩余的点按极角排序  计算后，第4维度存储角度，第5维度存储极径

#计算极角，极径
def PolarCorr(p0, p1):
    #print(p0, p1)
    dy = p1[0, 1] - p0[0, 1]
    dx = p1[0, 0] - p0[0, 0]
    if dx == 0:
        dx = 0.001
    gradient = dy / dx
    #print(gradient)
    angle = np.arctan(gradient) * 180 /np.pi
    if angle < 0:
        angle = angle + 180
    #print(angle)
    radius = np.sqrt(np.square(dy) + np.square(dx))
    #print(radius)
    return angle, radius
#转换剩余点
p0 = Data[pidx].reshape(1, 5)
#print(type(Data))
ResData = ResData.reshape(-1, 5)
#print(p0)
#print(ResData)
for idx in range(len(ResData)):
    pi = ResData[idx].reshape(1, 5)
    ang, rad = PolarCorr(p0, pi)
    ResData[idx, 3] = ang
    ResData[idx, 4] = rad
ResData = ResData[np.argsort(ResData[:, 3])]#排序来加快寻找
print('排序后\n', ResData, '\n')

#画图调试，查看每个点角度
x0 = p0[0, 0]
y0 = p0[0, 1]
for i in range(len(ResData)):
    xi = ResData[i, 0]
    yi = ResData[i, 1]
    x = np.linspace(x0, xi)
    y = np.linspace(y0, yi)
    plt.plot(x, y)
    plt.annotate(r'%s' %str(ResData[i, 3]),xy=(xi, yi), xycoords='data',\
xytext=(-20, +20), textcoords='offset points', fontsize=12,\
 arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.1"))
plt.show()
idy = 0
idx = 0
while idy < len(ResData) - 1:
    if ResData[idy, 3] == ResData[idy+1, 3]:
        if ResData[idy, 4] > ResData[idy+1, 4]:
            ResData = np.delete(Data, idy, axis = 0)
        else: 
            ResData = np.delete(Data, idy+1, axis = 0)
    idy = idy + 1
print('删除相同角之后：\n', ResData)
```

```python
##定义点和三角形的关系
def Point3_(p1, p2, p3):
    #print(i, data[i, :])
    #print(j, data[j, :])
   # print(p1, p2, p3)
    k = (p1[0, 1] - p2[0, 1])/(p1[0, 0] - p2[0, 0])
    re = k * (p3[0, 0] - p2[0, 0]) + (p2[0, 1] - p3[0, 1])
    return re
def Point2Tria(p1, p2, p3, p4):#p4是测试点
    if Point3_(p1, p2, p4)*Point3_(p1, p2, p3)>=0 and Point3_(p1, p3, p4)*Point3_(p1, p3, p2)>=0 \
        and Point3_(p2, p3, p4)*Point3_(p2, p3, p1)>=0:
        return True
    return False

#入结果栈
result = [p0, ResData[0], ResData[1]]
#剩余点
ResData = ResData[range(2, len(ResData))]
print('三点入栈后，剩余的元素:\n', ResData, '\n')

#开始筛选
for subscript in range(len(ResData)):
    pi = ResData[subscript].reshape(1, 5)
    ires = - 1
    while ires >= -1*len(result)+1:
        top = result[ires].reshape(1, 5)
        nextop = result[ires - 1].reshape(1, 5)
        if Point2Tria(p0, pi, nextop, top):
            result.pop()
            ires = -1
        else:
            result.append(pi)
            break
result = np.array(result)
print('结果数组:\n', result)
```
